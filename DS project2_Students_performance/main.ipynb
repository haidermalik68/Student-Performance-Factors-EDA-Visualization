```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.columns
```

```python
sns.lineplot(data=df, x='Hours_Studied', y='Exam_Score',hue='Gender', palette='Set1') 
                

plt.title('Exam Score vs Study Hours')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')

plt.show()
```

```python
sns.regplot(data=df, x='Hours_Studied', y='Exam_Score', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Exam Score vs Study Hours (Regression)')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

* There is a weak positive correlation between hours studied and exam score , however the relationship is not strongly significant, suggesting that other factors also play a major role in determining student performance *  

# ATTENDANCE  

```python
sns.regplot(data=df, x='Attendance', y='Exam_Score', scatter_kws={'alpha':0.9,'color':'gray'}, line_kws={'color':'red'}, ci=None,marker='x')
plt.title('Exam Score vs Attendance ')
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.show()
```

* There is a weak positive correlation between Attendance and exam score,suggesting that other factors also play a major role in determining student performance

# sleep hours

```python
sns.lineplot(data=df, x='Sleep_Hours', y='Exam_Score',hue='Gender', palette='Set2')
plt.title('Exam Score vs Sleep Hours ')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

*  the above graph shows that there is no clear relationship between sleep hours and exam score, as the score is not increasing or decreasing properly.

# Internet acces

```python
sns.boxplot(data=df, x='Internet_Access', y='Exam_Score', hue='Gender', palette='Set1')
plt.title('Exam Score vs Internet Access')
plt.xlabel('Internet Access')
plt.ylabel('Exam Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

* the box plot shows that students with internet access perform similarly to those without, with a slightly higher median score for those with access. The distribution of scores is also similar between the two groups, suggesting that internet access may not be a significant factor in exam performance for this dataset.

# Teacher quality
